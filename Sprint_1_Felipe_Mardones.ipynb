```python
class Producto:
# Esta clase representa los atributos que tiene un producto en el inventario, es la clase Producto (Debe comenzar con mayúscula por buenas prácticas).
    def __init__(self, id, nombre, descripcion, cantidad, precio):
        self.id = id
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

    # Métodos para obtener  atributos del producto, usamos self para poder acceder al atributo deseado en este caso todos los definidos anteriormente.

    # Permite retornar el id del producto.
    def obtener_id(self):
        return self.id

    # Permite retornar el nombre del producto.
    def obtener_nombre(self):
        return self.nombre

    # Permite retornar la descripción del producto.
    def obtener_descripcion(self):
        return self.descripcion

    # Permite retornar la cantidad del producto.
    def obtener_cantidad(self):
        return self.cantidad

    # Permite retornar el precio del producto.
    def obtener_precio(self):
        return self.precio

    # Métodos para modificar los atributos del producto
    # En este caso usamos self.atributo = nuevo_atributo para modificar el atributo.

    # Permite modificar el nombre de un producto por un nuevo valor.
    def modificar_nombre(self, nuevo_nombre):
        self.nombre = nuevo_nombre

    # Permite modificar la descripción de un producto por un nuevo valor.
    def modificar_descripcion(self, nueva_descripcion):
        self.descripcion = nueva_descripcion

    # Permite modificar la cantidad de un producto por un nuevo valor.
    def modificar_cantidad(self, nueva_cantidad):
        self.cantidad = nueva_cantidad

    # Permite modificar el precio de un producto por un nuevo valor.
    def modificar_precio(self, nuevo_precio):
        self.precio = nuevo_precio

# Esta clase representa el inventario de productos.
# Es la clase Inventario (Debe comenzar con mayúscula por buenas prácticas).

# Inicializa un nuevo objeto Inventario con un diccionario vacío para almacenar los productos y sus atributos.
class Inventario:
    def __init__(self):
        self.productos = {}

    # Los siguientes métodos permiten agregar, eliminar, actualizar, buscar y listar productos en el inventario.
    # Se usa if con in y not in para verificar si un elemento está presente en el diccionario, tambien se usa else y for con similar estructura.

    # Este método verifica si el producto existe en el diccionario self.productos = {} y sino lo agrega al diccionario.
    # Uso producto.id como key y la entrada producto es el valor, pude haber puesto cualquier cosa en producto.id pero debe ser algo distintivo.
    def agregar_producto(self, producto):
        if producto.id not in self.productos:
            self.productos[producto.id] = producto
            print("Producto agregado al inventario satisfactoriamente.")
        else:
            print("Producto ya presente en el inventario.")

    # Este método verifica si el producto esta en el diccionario y lo elimina y si no devuelve un mensaje predeterminado.
    def eliminar_producto(self, id_producto):
        if id_producto in self.productos:
            del self.productos[id_producto]
            print("Producto eliminado del inventario exitosamente.")
        else:
            print("No se pudo encontrar producto con ese ID en el inventario.")

    # Este método permite verificar si el producto con el ID especificado se encuentra en el inventario, actualiza sus atributos según los valores proporcionados.
    # Cualquier argumento opcional que no se especifique permanecerá sin cambios.
    # Imprime un mensaje indicando si la actualización se realizó exitosamente o si no se encontró ningún producto con el ID dado.
    def actualizar_producto(self, id_producto, nombre=None, descripcion=None, cantidad=None, precio=None):
        if id_producto in self.productos:
            producto = self.productos[id_producto]
            if nombre is not None:
                producto.modificar_nombre(nombre)
            if descripcion is not None:
                producto.modificar_descripcion(descripcion)
            if cantidad is not None:
                producto.modificar_cantidad(cantidad)
            if precio is not None:
                producto.modificar_precio(precio)
            print("Información del producto actualizada exitosamente.")
        else:
            print("No se pudo encontrar producto con ese ID en el inventario.")

    # Este método busca un producto en el inventario utilizando su ID como clave.
    # Si el producto se encuentra, devuelve el objeto Producto correspondiente.
    # Si no se encuentra ningún producto con el ID especificado, se imprime un mensaje predeterminado.
    def buscar_producto_por_id(self, id_producto):
        if id_producto in self.productos:
            return self.productos[id_producto]
        else:
            print("No se pudo encontrar producto con ese ID en el inventario.")

    # Este método imprime en pantalla la lista de todos los productos en el inventario, así como el total de productos y el valor total del inventario.
    def listar_productos(self):
            # Inicializa variables para almacenar el total de productos y el valor total del inventario.
            total_productos = 0
            valor_total = 0

            # Imprimir encabezado del listado de productos.
            print(" Productos en Inventario:")
            print("------------------------------------")

            # Iterar sobre cada producto en el inventario.
            for producto in self.productos.values():
                # Imprimir los atributos del producto.
                print("ID:", producto.id)
                print("Nombre:", producto.nombre)
                print("Descripción:", producto.descripcion)
                print("Cantidad:", producto.cantidad)
                print("Precio:", producto.precio)
                print("--------------------")

                # Sumar cantidad de productos
                total_productos = total_productos + producto.cantidad

                # Calcular valor total del producto
                valor_total = valor_total + (producto.cantidad * producto.precio)

            # Imprimir el total de productos y el valor total del inventario.
            print("Cantidad Total de Productos:", total_productos)
            print("Valor Total del Inventario:", valor_total)

# Método para imprimir un menu con opciones utilizando funciones anteriormente creadas.
# Usa input para elegir la opción deseada.
def mostrar_menu():
    print("Menú:")
    print("1. Agregar producto")
    print("2. Eliminar producto")
    print("3. Actualizar producto")
    print("4. Listar productos")
    print("5. Listado de productos y generar informe de totales")
    print("6. Salir")
    opcion = input("Seleccione una opción: ")
    return opcion

# Función principal que controla la interacción del usuario con el inventario.
# La finalidad de esta función es mostrar el menú de opciones creado al usuario y que le permita ejecutar la opción seleccionada.
def main():
    # Crea una nueva instancia de la clase Inventario.
    # Esta instancia se utilizará para trabajar con el inventario de productos.
    # Al llamar a la clase Inventario sin argumentos, se inicializa con un diccionario vacío para almacenar los productos.
    inventario = Inventario()

    # El bucle while True crea un ciclo infinito, lo que significa que se ejecutará indefinidamente hasta que se encuentre una condición de salida o se interrumpa.
    # Usamos break para corta la condición True y que no se cumpla para luego usar else para manejar las opciones no válidas que no están contempladas en el menú.
    # While Muestra el menú de opciones al usuario y espera a que el usuario seleccione una opción.
    # Try and except manejan el caso en el que el usuario ingresa un valor no válido durante la creación de un nuevo producto.
    while True:
        opcion = mostrar_menu()

        if opcion == "1":
            try:
                id_producto = int(input("Ingrese el ID del producto: "))
                nombre_producto = input("Ingrese el nombre del producto: ")
                descripcion_producto = input("Ingrese la descripción del producto: ")
                cantidad_producto = int(input("Ingrese la cantidad del producto: "))
                precio_producto = float(input("Ingrese el precio del producto: "))
                nuevo_producto = Producto(id_producto, nombre_producto, descripcion_producto, cantidad_producto, precio_producto)
                inventario.agregar_producto(nuevo_producto)
            except ValueError:
                print("Error: Entrada inválida. Por favor, ingrese un número válido.")

        elif opcion == "2":
            try:
                id_producto = int(input("Ingrese el ID del producto a eliminar: "))
                inventario.eliminar_producto(id_producto)
            except ValueError:
                print("Error: Entrada inválida. Por favor, ingrese un número válido.")

        elif opcion == "3":
            try:
                id_producto = int(input("Ingrese el ID del producto a actualizar: "))
                nombre = input("Ingrese el nuevo nombre del producto (o presione Enter para mantener el actual): ")
                descripcion = input("Ingrese la nueva descripción del producto (o presione Enter para mantener la actual): ")
                cantidad = int(input("Ingrese la nueva cantidad del producto (o presione Enter para mantener la actual): "))
                precio = float(input("Ingrese el nuevo precio del producto (o presione Enter para mantener el actual): "))
                inventario.actualizar_producto(id_producto, nombre, descripcion, cantidad, precio)
            except ValueError:
                print("Error: Entrada inválida. Por favor, ingrese un número válido.")

        elif opcion == "4":
            inventario.listar_productos()

        elif opcion == "5":
            if inventario.productos:
                print("Listado de productos e informe de totales:")
                print("--------------------------------------------")
                total_productos = 0
                valor_total = 0

                for producto in inventario.productos.values():
                    print("ID:", producto.id)
                    print("Nombre:", producto.nombre)
                    print("Descripción:", producto.descripcion)
                    print("Cantidad:", producto.cantidad)
                    print("Precio:", producto.precio)
                    print("--------------------")

                    total_productos += producto.cantidad
                    valor_total += producto.cantidad * producto.precio

                print("Cantidad Total de Productos:", total_productos)
                print("Valor Total del Inventario:", valor_total)
            else:
              print("No hay productos en el inventario para generar un informe de totales.")

        elif opcion == "6":
            print("Saliendo del programa...")
            break

        else:
            print("Opción no válida. Por favor, seleccione una opción válida del menú.")

# Este bloque condicional verifica si el script se está ejecutando como el programa principal. Si es así, llama a la función main() para iniciar la ejecución del programa.
# Esto permite que el código sea reutilizable como un módulo importable en otros scripts sin ejecutar automáticamente la función main().
if __name__ == "__main__":
    main()

# Fin Código.
# Link GitHube
https://github.com/Felipecl22/BootCamp/blob/2fe227ad4eebc41e2f75aa3be337cb372a6a40c4/Sprint_1_Felipe_Mardones.ipynb
```
